# ‚öñ Normalizacion (Escalado)

Escalar todas las variables num√©ricas para que tengan un rango comparable.

Normalizar usando StandardScaler de scikit-learn, que estandariza los datos restando la media y dividiendo por la desviaci√≥n est√°ndar. Ideal para K-Means.

```python
# Vamos a normalizar solo las variables numericas
# Esto crea un nuevo DataFrame solo con las columnas que queremos escalar
numeric_columns = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Aplicar el escalado con StandardScaler
scaler = StandardScaler() # Invoco la funcion
scaled_numeric_columns = scaler.fit_transform(numeric_columns)
# fit_transform calcula la media y desviaci√≥n estandar, luego transforma los datos
```

```python
# Creo un DataFrame con los datos ya escalados(normalizados)
df_scaled = pd.DataFrame(scaled_numeric_columns, columns=numeric_columns.columns) # asignando los nombres de las columnas originales
# Esto nos muestra las primeras filas de los datos normalizados para verificar el resultado
print(df_scaled.head())
```

```python
df_scaled = pd.DataFrame(scaled_numeric_columns, columns=numeric_columns.columns)
print(df_scaled.tail())
```

Conclusi√≥n: Despu√©s de escalar los datos, las columnas Age, Annual Income (k$) y Spending Score (1-100) tienen una media de 0 y una desviaci√≥n est√°ndar de 1. 

Esto significa que:

üî∫ Los valores positivos indican que el dato est√° por encima de la media.

üîª Los valores negativos indican que el dato est√° por debajo de la media.

üîµ Los valores cercanos a 0 representan datos muy pr√≥ximos a la media.

Al realizar este proceso, logramos que todas las variables tengan el mismo peso al momento de aplicar algoritmos de agrupamiento como K-Means, evitando que variables con diferentes escalas(como ingresos altos frente a edades) dominen el an√°lisis. 





# üí™üèΩ M√©todo del codo

```python
# Usamos el metodo del codo para encontrar un k optimo (k es la cantidad de grupos o clusteres que queremos formar)
# Creamos una lista vacia para guardar la inercia
# La inercia mide qu√© tan cerca est√°n los puntos de sus centroides dentro de cada cl√∫ster. Cuanto menor, mejor.
inertia = []
```

```python
# Bucle para probar varios valores de K (cantidad de cl√∫steres), t√≠picamente entre 1 y 10:
for k in range(1, 11):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    # Guardo dentro de la variable kmeans el modelo para luego usarlo con .fit() y entrenarlo con mis datos
    # random_state = 42, Sirve para que el resultado sea reproducible. Se puede usar cualquier n√∫mero (por convenci√≥n se suele usar 42). 
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), inertia, marker='o')
# marker='o' dibuja puntos
plt.title('M√©todo del Codo')
plt.xlabel('N√∫mero de Cl√∫steres (K)')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()
```

Conclusi√≥n: El codo se nota claramente en K = 4.
Despu√©s de ese punto, la inercia sigue bajando, pero cada vez menos, lo que indica que agregar m√°s cl√∫steres(grupos) no mejora tanto el agrupamiento de clientes.

# üí°K-Means

```python
# Inicializacion KMeans con K=4
Kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
```

```python
# fit_predict: entrena el modelo sobre los datos escalados y devuelve un array con la etiqueta de cada cliente
labels = kmeans.fit_predict(df_scaled)
```

```python
# Ahora cada fila (cliente) tiene asignado un grupo, lo que permite analizar y describir cada segmento.
df['Cluster'] = labels
```

```python
# Ver las primeras filas con su cluster asignado
print(df.head())
```

#### Distribuci√≥n de g√©nero en cada cl√∫ster

```python
# Conteo de g√©nero por cluster
gender_dist = df.groupby(["Cluster", "Gender"]).size().unstack(fill_value=0)
# Calcular % dentro de cada cluster
gender_percent = gender_dist.div(gender_dist.sum(axis=1), axis=0) * 100
print(gender_percent.round(1))
```

Obtenemos que g√©nero predomina el cl√∫ster y con esto orientar campa√±as efectivas
